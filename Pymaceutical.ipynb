```python
#imports

import pandas as pd
import numpy  as np

#plotting library
import matplotlib.pyplot as plt
```

```python
#source data
mouseDrugPd = pd.read_csv("raw_data/mouse_drug_data.csv")
clinicalPd = pd.read_csv("raw_data/clinicaltrial_data.csv")

DrugsOfConcern = ["Capomulin", "Infubinol", "Ketapril","Placebo"]
```

```python
#intrnal data format remove this
print(f"MouseDrugPd = {mousePd.head(3)}")
print("=========================================")
print(f"clinicalPd = {clinicalPd.head(3)}")
```

```python
##len (clinicalPd['Mouse ID'].unique()) == len (mouseDrugPd['Mouse ID'].unique()) # = 249 
```

```python
#Join tables
#Both data frames are index by sequerial number
#merge bot dataframes
#re-index both 'Mouse ID' and allow duplicate keys
#it appears each rat got one drug
mPd = mouseDrugPd
cPd = clinicalPd.set_index('Mouse ID')
data = mPd.merge(clinicalPd)
data.set_index('Mouse ID',inplace=True)

data.loc['b128']
```

```python
volumeOverTimeDf = data.reset_index()
volumeOverTimeDf.sort_values(['Timepoint','Tumor Volume (mm3)']).head(5)
```

```python
#Group showing how each drug 
drug_Group = data.groupby(['Drug','Timepoint','Tumor Volume (mm3)']) 


drug_Group.sum()
```

```python
#columns of drug versus index of timepoints
timeVsDrugDict = {}
timePoints = []

timePointVsDrugGroup = data.groupby(['Timepoint'])
totalDrugList = set(data["Drug"])

#pre-seed with empty list
for drugName in totalDrugList:
   timeVsDrugDict[drugName] = []; 


for timeP, r_df in timePointVsDrugGroup:
    timePoints.append(timeP)
    for drugName in totalDrugList: 
       timeVsDrugDict[drugName].append(np.mean(r_df.loc[r_df['Drug'] == drugName,'Tumor Volume (mm3)']))
    
    
        

timeVsDrugDict['Timepoint'] = timePoints 
timeVsDrug_df = pd.DataFrame.from_dict(timeVsDrugDict)

timeVsDrug_df.set_index("Timepoint", inplace=True)

timeVsDrug_df
```
